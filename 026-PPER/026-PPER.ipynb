## 026-PPER Patial Permutations 部分順列
### 課題背景
- 似ている種は多くの遺伝子を共有しているが、完全に同じというよりは、改変もある
- 遺伝子の並びを解析することで二つのゲノムを比較することができ、どのゲノム再編成が遺伝子を分離させたかを知ることができる
- 018-PERMでは、順列を使うことでシンテニーブロックの並び順をモデル化できた。
-  しかし、二つのゲノムはそれぞれ独自の進化経路をたどってきたため、すべての遺伝子を共有しているわけではない。その結果、部分的な遺伝子順序付けの概念を定量化するために、部分順列を考える必要性がある

### 課題
- 部分順列とは、n 個の要素の中から k 個取り出して並べるということである
- 例えば、1 - 8 の数字から3 つ取り出して並べるとき、(5, 7, 2)のような順列が得られる
- Given 
  - 正の整数 n と k
    - 0 < n <= 100, 0 < k <= 10
- Return
  - P(n, k) を 1,000,000 で割ったときの余り

### 回答
- itertools の permutationsを使う

```python
n = 21
k = 7
```

```python
# 実際に順列を生成しているのでめちゃくちゃ遅い
from itertools import permutations
count = 0
for p in permutations(range(n), k):
    count += 1
print(count % 1000000)
```

```python
# ただ計算するだけなので、P(n, k) = n! / (n-k)!
from math import factorial
partial_permutations = (factorial(n) // factorial(n - k)) % 1000000
print(partial_permutations)
```

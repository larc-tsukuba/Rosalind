# 047-MOTZ モツキン数とRNA二次構造

RNAのすべての塩基が必ずペアになる場合（完全マッチング）：  
→ カタラン数 （https://github.com/larc-tsukuba/Rosalind/blob/main/032-CAT/032_cat.ipynb）

今回は、 **ペアにならない塩基があってもよい** 場合を考える：  
→ モツキン数

実際のRNAは塩基対を作るものもあれば作らないものもあるので、カタラン数よりもモツキン数のほうが現実を反映している。  

```python
rna = "AUAU"
```

```python
MOD  = 1_000_000
PAIRABLE = {
    ("A", "U"),
    ("U", "A"),
    ("C", "G"),
    ("G", "C"),
}


n = len(rna)

dp = [[0] * n for _ in range(n)]

# 長さ1でも1通りのマッチングがあるとする
for i in range(n):
    dp[i][i] = 1


def get_dp(left: int, right: int) -> int:
    """
    Return the DP value for interval [left, right].
    An empty interval has exactly one valid matching.
    """
    if left > right:
        return 1
    return dp[left][right]

for interval_length in range(2, n + 1):
    for interval_left in range(n - interval_length + 1):
        interval_right = interval_left + interval_length - 1

        total_count = get_dp(interval_left + 1, interval_right)

        for partner_index  in range(interval_left + 1, interval_right + 1):
            if (rna[interval_left], rna[partner_index]) in PAIRABLE:
                inside = get_dp(interval_left + 1, partner_index - 1)
                outside = get_dp(partner_index + 1, interval_right)
                total_count += inside * outside

        dp[interval_left][interval_right] = total_count % MOD

print(dp[0][n - 1])
```

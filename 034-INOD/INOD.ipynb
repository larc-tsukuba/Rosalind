## 問題

二分木とは，各ノードの次数が最大 3 の木である。二分木は系統樹（phylogeny）の構築における主要な道具となる。

**根付き木**とは，1 つのノード（根）を木の頂点として区別した木である。グラフ理論の基本的な事実として，任意の 2 つのノードの間にはちょうど 1 本のパスが存在する。したがって根付き木では，任意のノード \(v\) は一意の親ノードを持つ。この親とは，\(v\) から根へ向かうパスに含まれる \(\{v, w\}\) の形の隣接ノード \(w\) である。

\(v\) に隣接し，親以外のノード \(x\) は **子ノード** と呼ばれる。1 つのノードは複数の子を持つことができる。  
根付き木には根から葉へと向かう階層構造があり，無向グラフとして表された根付き木を，しばしば **親から子へ向かう有向グラフ** とみなすことができる。Rosalind の問題ツリーもこの考え方を用いている。

---

## 二分木の定義

二分木には次数 2 のノードを含むことがあるが，**根なし二分木（unrooted binary tree）** は特に「すべての内部ノードの次数が 3」であると定義される。

一方，**根付き二分木（rooted binary tree）** は「根のみが次数 2 を持ち，その他の内部ノードはすべて次数 3」である。

---

## 入力

正の整数  
\[
n  (3 <= n <= 10000)
\]

---

## 出力

葉が \(n\) 個ある **任意の根なし二分木** における **内部ノードの個数** を求めよ。


```python
def result(n):
    return n -2 
```

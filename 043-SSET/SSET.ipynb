# Counting Subsets

### 生物学的背景
形質の1つとしてSNVがある．  
SNVは読み替えるとon/offのスイッチと読み替えることもできる．  

例
WT:  AAAAAAAA  
v1:  AAAAAA**G**A  
v2:  A**G**AAA**G**A  

v1はSNVは7番目がon, v2はSNVは3番目と7番目がonである．
このようにon/offのスイッチがn個と考えると，ある生物の配列を1つの集合で表すことができる．

## 問題

**集合**とは、要素と呼ばれる対象のゆるやかな集まりを表す数学用語です。  
集合の例としては、`{月, 太陽, Wilford Brimley}` や、すべての実数を含む集合 `R` があります。  
要素をまったく含まない **空集合** もあり、`∅` または `{}` で表します。

2つの集合は、**同じ要素を含んでいるとき等しい**といいます。  
つまり、順列とは異なり、**集合では要素の順序は重要ではありません**。  
たとえば、

`{月, 太陽, Wilford Brimley}`

は

`{Wilford Brimley, 月, 太陽}`

と同じ集合です。

また、集合には **重複した要素を含めることはできません**。  
したがって、

`{Wilford Brimley, 太陽, 太陽}`

は集合ではありません。  
私たちはすでに、グラフの辺を表すために 2 要素の集合を使ってきました。

集合 `A` が `B` の **部分集合** であるとは、`A` のすべての要素が `B` にも含まれていることをいい、`A ⊆ B` と書きます。  
たとえば、

`{太陽, 月} ⊆ {太陽, 月, Wilford Brimley}`

です。  
また、空集合 `∅` は、**あらゆる集合の部分集合**です（自分自身も含みます）。

生物学の導入で示したように、部分集合を使って「ある形質をもつ分類群の集まり」を表すことができます。  
しかし、その応用は無数にあります。  
たとえば確率論では、事象を「すべての可能な結果からなる集合の部分集合」として定義できます。

最初の問いとして、与えられた集合に対して **取りうる部分集合の総数** を数えます。

### 与えられるもの

正の整数 `n`（`n ≤ 1000`）。

### 求めるもの

集合 `{1, 2, …, n}` の部分集合の総数を `1,000,000` で割った余りとして返してください。

いわゆるべき集合を求める問題．

```python
def answer(integer):
    return 2**integer  % 1000000
```

```python
answer(3)
```

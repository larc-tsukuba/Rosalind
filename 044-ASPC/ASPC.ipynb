````python
MOD = 1_000_000
n = 6
m = 3
# combination
from math import factorial

def comb(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))

ans = 0
for k in range(m, n+1):
    ans += comb(n, k) % MOD
print(ans % MOD)
````
